```python
import pandas as pd
# Read in data and examine first 10 rows
songs = pd.read_csv('bpm_songs.csv')
#print (songs)
#flights.head(10)
# Import the libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# seaborn histogram
sns.distplot(songs['tempo'], hist=True, kde=False, 
             bins=int(98/10), color = 'blue',
             hist_kws={'edgecolor':'black'})
# Add labels
plt.title('Histogram of Arrival Delays')
plt.xlabel('Delay (min)')
plt.ylabel('Flights')
```

```python
stis = ['mot', 'mid', 'rel']
```

```python
sns.set(color_codes=True)

# Iterate through the five airlines
for sti in stis:
    # Subset to the airline
    subset = songs[songs['sti'] == sti]
    # Draw the density plot
    color = 'black'
    if sti == 'mot':
        color = '#D8D8D8'
    if sti == 'mid':
        color = 'gray'
    sns.distplot(subset['tempo'], hist = False, kde = True,
                 kde_kws = {'shade': True, 'linewidth': 3},
                 color = color,
                 label = sti)
# Plot formatting
plt.legend(prop={'size': 16}, title = 'Tempo')
#plt.title('Density Plot with Multiple Stimulus')
plt.xlabel('Tempo')
plt.ylabel('Density')
#no grid
plt.grid(False)
```

```python
# Make a separate list for each airline
x1 = list(songs[songs['mode'] == 'minor']['sti'])
x2 = list(songs[songs['mode'] == 'major']['sti'])

# Assign colors for each airline and the names
colors = ['#E69F00', '#56B4E9']
names = ['minor', 'major']
         
# Stacked histogram with multiple airlines
plt.hist([x1, x2], stacked=True,
         normed=True, color = colors, label=names)
# Plot formatting
plt.legend()
plt.xlabel('Stimulus')
plt.ylabel('Normalized songs')
#plt.title('Stacked Histogram with Modes')
```

```python
# Make a separate list for each airline
x1 = list(songs[songs['articulation'] == 'Legato']['sti'])
x2 = list(songs[songs['articulation'] == 'Staccato']['sti'])

# Assign colors for each airline and the names
colors = ['#E69F00', '#56B4E9']
names = ['legato', 'staccato']
         
# Stacked histogram with multiple airlines
plt.hist([x1, x2], stacked=True,
         normed=True, color = colors, label=names)
# Plot formatting
plt.legend()
plt.xlabel('Stimulus')
plt.ylabel('Normalized songs')
#plt.title('Stacked Histogram with Articulation')
```
